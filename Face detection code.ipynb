```python
print(detections[0])
print(detections[0].left())
print(detections[0].top())
print(detections[0].right())
print(detections[0].bottom())
cv2.rectangle(image, (detections[0].left(), detections[0].top()), (detections[0].right(), detections[0].bottom()), (0, 255, 255), 2)
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```

```python
for face in detections:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l, t), (r, b), (0, 255, 255), 2)
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```

```python
cnn_detector = dlib.cnn_face_detection_model_v1('C:\\Users\\KIIT\\Acmegrade Machine Learning\\Project\\PRJFace Detection Project\\mmod_human_face_detector.dat')
display (cnn_detector)
```

```python
detections = cnn_detector(image, 1)
for face in detections:
  l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
  print(c)
  cv2.rectangle(image, (l, t), (r, b), (255, 255, 0), 2)
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```

```python
#Haarcascade

image = cv2.imread('C:\\Users\\KIIT\\Acmegrade Machine Learning\\Project\\PRJFace Detection Project\\people3.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
haarcascade_detector = cv2.CascadeClassifier('C:\\Users\\KIIT\\Acmegrade Machine Learning\\Project\\PRJFace Detection Project\\haarcascade_frontalface_default.xml')
detections = haarcascade_detector.detectMultiScale(image_gray, scaleFactor = 1.001, minNeighbors=5, minSize = (5,5))
for (x, y, w, h) in detections:
  cv2.rectangle(image, (x, y), (x + w, y + h), (0,255,0), 2)
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```
